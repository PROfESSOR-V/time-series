```python
def optimize_memory(df):
    #downcast numeric columns to reduce RAM and improve training throughput.
    for col in df.columns:
        if pd.api.types.is_float_dtype(df[col]):
            df[col] = pd.to_numeric(df[col], downcast='float')
        elif pd.api.types.is_integer_dtype(df[col]):
            df[col] = pd.to_numeric(df[col], downcast='integer')
    return df
```

```python
def decode_ordinals(df):
    #Decode exponentially encoded sensor features.
    df['X3_ord'] = np.round(np.log(np.maximum(df['X3'].values, 1e-10))).astype(np.int16)
    df['X4_ord'] = np.round(np.log(np.maximum(df['X4'].values, 1e-10))).astype(np.int16)
    df['X5_ord'] = np.round(np.exp(df['X5'].values)).astype(np.int16)
    return df
```

```python
def add_zone_features(df):
    #Add deterministic anomaly zone indicators - KEY NEW FEATURES.
    X3 = df['X3_ord']; X4 = df['X4_ord']; X5 = df['X5_ord']
    osum = X3 + X4

    df['zone_certain'] = ((X3 >= 60) | (X4 >= 64) | ((X3 >= 50) & (X4 >= 30))).astype(np.int8)
    df['zone_high'] = ((X3 >= 40) & (X4 >= 40)).astype(np.int8)
    df['zone_moderate'] = ((osum >= 50) & (X5 >= 17)).astype(np.int8)
    df['zone_elevated'] = (((X3 >= 35) | (X4 >= 34)) & (X5 >= 17)).astype(np.int8)
    df['zone_score'] = (df['zone_certain']*4 + df['zone_high']*3 +
                        df['zone_moderate']*2 + df['zone_elevated']).astype(np.int8)
    df['zone_safe'] = ((X3 == 0) & (X4 == 0) & (X5 <= 1)).astype(np.int8)
    return df
```

```python
def add_ordinal_features(df):
    #Add ordinal indicators and interactions.
    X3 = df['X3_ord']; X4 = df['X4_ord']; X5 = df['X5_ord']

    # Ordinal indicators
    for col, name, thresholds in [
        (X3, 'X3', [20, 30, 35, 40, 50, 60]),
        (X4, 'X4', [20, 30, 34, 40, 50, 64]),
    ]:
        for t in thresholds:
            df[f'{name}_ge{t}'] = (col >= t).astype(np.int8)

    df['X5_peak'] = (X5 == 18).astype(np.int8)
    df['X5_ge17'] = (X5 >= 17).astype(np.int8)
    df['X5_17_19'] = (X5.isin([17,18,19])).astype(np.int8)

    # Risk level
    df['risk_level'] = (df['X3_ge35'] + df['X3_ge50'] + df['X3_ge60'] +
                        df['X4_ge34'] + df['X4_ge50'] + df['X4_ge64'] +
                        df['X5_ge17']).astype(np.int8)

    # Pairwise interactions
    df['X3X4_sum'] = (X3 + X4).astype(np.int16)
    df['X3X4_diff'] = (X3 - X4).astype(np.int16)
    df['X3X4_absdiff'] = np.abs(df['X3X4_diff']).astype(np.int16)
    df['X3X4_product'] = (X3 * X4).astype(np.int32)
    df['X3X4_max'] = np.maximum(X3, X4).astype(np.int16)
    df['X3X4_min'] = np.minimum(X3, X4).astype(np.int16)
    df['X3X5_sum'] = (X3 + X5).astype(np.int16)
    df['X3X5_product'] = (X3 * X5).astype(np.int32)
    df['X4X5_sum'] = (X4 + X5).astype(np.int16)
    df['X4X5_product'] = (X4 * X5).astype(np.int32)
    df['X4X5_diff'] = (X4 - X5).astype(np.int16)
    df['X3X4X5_sum'] = (X3 + X4 + X5).astype(np.int16)
    df['X3X4X5_max'] = np.maximum(np.maximum(X3, X4), X5).astype(np.int16)
    df['X3X4X5_product'] = (X3 * X4 * np.maximum(X5, 1)).astype(np.int64)

    # Squared terms
    df['X3_sq'] = (X3**2).astype(np.int32)
    df['X4_sq'] = (X4**2).astype(np.int32)
    df['X3X4_sum_sq'] = (df['X3X4_sum']**2).astype(np.int32)

    # Ratios
    df['X3_X4_ratio'] = (X3 / (X4 + 1)).astype(np.float32)
    df['X4_X3_ratio'] = (X4 / (X3 + 1)).astype(np.float32)
    return df
```

```python
def add_continuous_features(df):
    #Add continuous sensor interactions.
    X1 = df['X1']; X2 = df['X2']
    df['X1X2_ratio'] = (X1 / X2.clip(lower=4.5)).astype(np.float32)
    df['X1X2_product'] = (X1 * X2).astype(np.float32)
    df['X1X2_diff'] = (X1 - X2).astype(np.float32)
    df['X1_sq'] = (X1**2).astype(np.float32)
    df['X2_sq'] = (X2**2).astype(np.float32)
    df['X1_log'] = np.log1p(X1).astype(np.float32)
    df['X2_log'] = np.log1p(X2).astype(np.float32)
    df['X2_dev'] = (X2 - X2.mean()).astype(np.float32)
    df['X2_neg_dev'] = (-df['X2_dev']).clip(lower=0).astype(np.float32)

    # Continuous × ordinal
    for oc in ['X3_ord', 'X4_ord', 'X5_ord']:
        df[f'X1_{oc}_prod'] = (X1 * df[oc]).astype(np.float32)
        df[f'X2_{oc}_prod'] = (X2 * df[oc]).astype(np.float32)
    return df
```

```python
def add_time_features(df):
    """Add temporal features."""
    df['dayofweek'] = df['Date'].dt.dayofweek.astype(np.int8)
    df['month'] = df['Date'].dt.month.astype(np.int8)
    df['day'] = df['Date'].dt.day.astype(np.int8)
    df['quarter'] = df['Date'].dt.quarter.astype(np.int8)
    df['dayofyear'] = df['Date'].dt.dayofyear.astype(np.int16)
    df['is_weekend'] = (df['dayofweek'] >= 5).astype(np.int8)
    df['month_sin'] = np.sin(2*np.pi*df['month']/12).astype(np.float32)
    df['month_cos'] = np.cos(2*np.pi*df['month']/12).astype(np.float32)
    df['dow_sin'] = np.sin(2*np.pi*df['dayofweek']/7).astype(np.float32)
    df['dow_cos'] = np.cos(2*np.pi*df['dayofweek']/7).astype(np.float32)
    return df
```

```python
def add_date_agg_features(df):
    #Add date-level aggregation features
    conts = ['X1', 'X2']
    ords = ['X3_ord', 'X4_ord', 'X5_ord']
    for col in conts + ords:
        grp = df.groupby('Date')[col]
        df[f'{col}_dmean'] = grp.transform('mean').astype(np.float32)
        df[f'{col}_dstd'] = grp.transform('std').fillna(0).astype(np.float32)
        df[f'{col}_dmax'] = grp.transform('max').astype(np.float32)
        df[f'{col}_vs_dmean'] = (df[col] - df[f'{col}_dmean']).astype(np.float32)

    df['date_count'] = df.groupby('Date')['X1'].transform('count').astype(np.int16)
    for ind in ['X3_ge35', 'X4_ge34', 'X5_ge17']:
        df[f'date_{ind}_count'] = df.groupby('Date')[ind].transform('sum').astype(np.int16)
        df[f'date_{ind}_prop'] = (df[f'date_{ind}_count'] / df['date_count']).astype(np.float32)

    df['date_risk_total'] = (df['date_X3_ge35_count'] + df['date_X4_ge34_count']).astype(np.int16)
    df['date_risk_prop'] = (df['date_risk_total'] / df['date_count']).astype(np.float32)
    df['date_X3_max'] = df.groupby('Date')['X3_ord'].transform('max').astype(np.int16)
    df['date_X4_max'] = df.groupby('Date')['X4_ord'].transform('max').astype(np.int16)

    for col in conts:
        df[f'{col}_drank'] = df.groupby('Date')[col].rank(pct=True).astype(np.float32)
    return df
```

```python
def add_freq_features(df):
    #Add frequency encoding features.
    for col in ['X3_ord', 'X4_ord', 'X5_ord']:
        freq = df[col].value_counts(normalize=True)
        df[f'{col}_freq'] = df[col].map(freq).astype(np.float32)

    pair_freq = df.groupby(['X3_ord','X4_ord']).size() / len(df)
    pair_map = pair_freq.to_dict()
    df['X3X4_pfreq'] = pd.Series(
        [pair_map.get((a,b), 0) for a,b in zip(df['X3_ord'], df['X4_ord'])],
        index=df.index, dtype=np.float32
    )
    return df
```

```python
def add_iso_features(df):
    #Add Isolation Forest anomaly scores
    feats = ['X1', 'X2', 'X3_ord', 'X4_ord', 'X5_ord']
    iso = IsolationForest(n_estimators=300, contamination=0.01, random_state=42, n_jobs=-1)
    iso.fit(df[feats].values)
    df['iso_score'] = iso.score_samples(df[feats].values).astype(np.float32)
    df['iso_anomaly'] = (iso.predict(df[feats].values) == -1).astype(np.int8)
    del iso; gc.collect()
    return df
```

```python
def add_target_encoding(train_df, test_df, te_cols, target_col='target', n_splits=5, smooth=20):
    #K-fold Bayesian-smoothed target encoding.
    global_mean = train_df[target_col].mean()
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=99)
    for col in te_cols:
        train_df[f'{col}_te'] = 0.0
        for tr_idx, va_idx in skf.split(train_df, train_df[target_col]):
            tr = train_df.iloc[tr_idx]
            stats = tr.groupby(col)[target_col].agg(['mean','count'])
            stats['s'] = (stats['count']*stats['mean'] + smooth*global_mean) / (stats['count'] + smooth)
            train_df.iloc[va_idx, train_df.columns.get_loc(f'{col}_te')] = (
                train_df.iloc[va_idx][col].map(stats['s']).fillna(global_mean).values
            )
        full_stats = train_df.groupby(col)[target_col].agg(['mean','count'])
        full_stats['s'] = (full_stats['count']*full_stats['mean'] + smooth*global_mean) / (full_stats['count'] + smooth)
        test_df[f'{col}_te'] = test_df[col].map(full_stats['s']).fillna(global_mean).astype(np.float32)
        train_df[f'{col}_te'] = train_df[f'{col}_te'].astype(np.float32)
    return train_df, test_df
```

```python
def build_features(train_df, test_df):
    #Complete feature engineering pipeline
    print("FEATURE ENGINEERING ON COMBINED DATA")
    if FAST_MODE:
        print("FAST_MODE=True -> skipping the slowest CPU-only feature blocks by default")

    train_df['_is_train'] = 1
    test_df['_is_train'] = 0
    combined = pd.concat([train_df, test_df], ignore_index=True)
    target = combined['target'].copy()

    print("  -> Decoding ordinals...")
    combined = decode_ordinals(combined)
    print("  -> Zone features...")
    combined = add_zone_features(combined)
    print("  -> Ordinal features...")
    combined = add_ordinal_features(combined)
    print("  -> Continuous features...")
    combined = add_continuous_features(combined)
    print("  -> Time features...")
    combined = add_time_features(combined)
    print("  -> Date aggregations...")
    combined = add_date_agg_features(combined)
    print("  -> Frequency encoding...")
    combined = add_freq_features(combined)
    if USE_ISO_FEATURES:
        print("  -> Isolation Forest...")
        combined = add_iso_features(combined)
    else:
        print("  -> Skipping Isolation Forest in FAST_MODE")

    # Split back
    print("  -> Splitting back...")
    train_mask = combined['_is_train'] == 1
    train_out = combined[train_mask].copy()
    test_out = combined[~train_mask].copy()
    train_out['target'] = target[train_mask].values
    del combined; gc.collect()

    # Target encoding (needs target, so done after split)
    if USE_TARGET_ENCODING:
        print("  -> Target encoding...")
        te_cols = ['X3_ord', 'X4_ord', 'X5_ord', 'X3X4_sum', 'X4X5_sum', 'X3X4_max']
        train_out, test_out = add_target_encoding(train_out, test_out, te_cols)
    else:
        print("  -> Skipping target encoding in FAST_MODE")

    train_out = optimize_memory(train_out)
    test_out = optimize_memory(test_out)

    print(f"  Train: {train_out.shape}, Test: {test_out.shape}")
    return train_out, test_out

```

#Architecture:
  - Rich engineered features with zone-based anomaly indicators
  - Fast default mode: GPU-first ensemble (XGB x2 + CatBoost)
  - Full mode available by setting FAST_MODE=False
  - Threshold optimization + rule-based post-processing

```python
# 2. FEATURE ENGINEERING
train_df, test_df = build_features(train_df, test_df)
```

```python
# 3. PREPARE FEATURES
drop_cols = ['target', 'Date', 'ID', 'X3', 'X4', 'X5', '_is_train']
features = [c for c in train_df.columns if c not in drop_cols]

for c in features:
    train_df[c] = train_df[c].fillna(0)
    test_df[c] = test_df[c].fillna(0)

X = train_df[features].values.astype(np.float32)
y = train_df['target'].values.astype(np.int8)
X_test = test_df[features].values.astype(np.float32)
```

```python
# Save zone columns for post-processing
train_zone_certain = train_df['zone_certain'].values
train_zone_safe = train_df['zone_safe'].values
test_zone_certain = test_df['zone_certain'].values
test_zone_safe = test_df['zone_safe'].values
test_X3_ord = test_df['X3_ord'].values
test_X4_ord = test_df['X4_ord'].values

neg = (y == 0).sum()
pos = (y == 1).sum()
scale_w = neg / pos
print(f"Features: {len(features)}")
print(f"Samples: {len(X)}, Pos: {pos} ({100*pos/len(X):.2f}%), Ratio: {scale_w:.1f}:1")
print(f"Feature names: {features[:10]}... (showing first 10)")
```

```python
def get_models(fold_seed):
    #Return list of (name, model) tuples. Uses GPU where available.
    if FAST_MODE:
        return [
            ('XGB_deep', XGBClassifier(
                scale_pos_weight=scale_w * 0.55,
                n_estimators=1400, max_depth=8, learning_rate=0.03,
                subsample=0.80, colsample_bytree=0.65, colsample_bylevel=0.8,
                min_child_weight=8, gamma=0.15, reg_alpha=0.3, reg_lambda=2.0,
                random_state=42 + fold_seed, n_jobs=-1,
                tree_method='hist', device='cuda', max_bin=256,
                eval_metric='aucpr', early_stopping_rounds=100,
            )),
            ('XGB_wide', XGBClassifier(
                scale_pos_weight=scale_w * 0.65,
                n_estimators=1100, max_depth=5, learning_rate=0.05,
                subsample=0.90, colsample_bytree=0.80,
                min_child_weight=18, gamma=0.05, reg_alpha=0.2, reg_lambda=1.5,
                random_state=100 + fold_seed, n_jobs=-1,
                tree_method='hist', device='cuda', max_bin=256,
                eval_metric='aucpr', early_stopping_rounds=100,
            )),
            ('CatBoost', CatBoostClassifier(
                auto_class_weights='Balanced',
                iterations=1200, learning_rate=0.05, depth=6,
                l2_leaf_reg=5.0, random_seed=400 + fold_seed,
                verbose=0, eval_metric='AUC', allow_writing_files=False,
                early_stopping_rounds=100, task_type='GPU', devices='0',
                border_count=128,
            )),
        ]

    models = []

    # Model 1: XGBoost Deep - captures complex interactions
    models.append(('XGB_deep', XGBClassifier(
        scale_pos_weight=scale_w * 0.5,
        n_estimators=2500, max_depth=9, learning_rate=0.02,
        subsample=0.7, colsample_bytree=0.5, colsample_bylevel=0.7,
        min_child_weight=5, gamma=0.2, reg_alpha=0.5, reg_lambda=3.0,
        random_state=42 + fold_seed, n_jobs=-1,
        tree_method='hist', device='cuda', max_bin=256,
        eval_metric='aucpr', early_stopping_rounds=150,
    )))

    # Model 2: XGBoost Wide - robust shallow learner, higher recall
    models.append(('XGB_wide', XGBClassifier(
        scale_pos_weight=scale_w * 0.7,
        n_estimators=2000, max_depth=5, learning_rate=0.03,
        subsample=0.85, colsample_bytree=0.7,
        min_child_weight=15, gamma=0.1, reg_alpha=0.3, reg_lambda=2.0,
        random_state=100 + fold_seed, n_jobs=-1,
        tree_method='hist', device='cuda', max_bin=256,
        eval_metric='aucpr', early_stopping_rounds=150,
    )))

    # Model 3: LightGBM - scale_pos_weight (is_unbalance collapses at iter 14)
    models.append(('LGB', LGBMClassifier(
        scale_pos_weight=scale_w * 0.5,
        n_estimators=2000, learning_rate=0.02,
        num_leaves=127, max_depth=8,
        subsample=0.75, subsample_freq=1, colsample_bytree=0.6,
        min_child_samples=20, reg_alpha=1.0, reg_lambda=3.0,
        random_state=200 + fold_seed, n_jobs=-1, verbose=-1,
    )))

    # Model 4: LightGBM GOSS - gradient-based sampling, focuses on hard examples
    models.append(('LGB_goss', LGBMClassifier(
        boosting_type='goss',
        is_unbalance=True,
        n_estimators=1500, learning_rate=0.03,
        num_leaves=127, max_depth=8,
        colsample_bytree=0.7,
        min_child_samples=20, reg_alpha=1.0, reg_lambda=3.0,
        random_state=300 + fold_seed, n_jobs=-1, verbose=-1,
    )))

    # Model 5: CatBoost - scale_pos_weight approach
    models.append(('CatBoost', CatBoostClassifier(
        scale_pos_weight=scale_w * 0.5,
        iterations=1500, learning_rate=0.03, depth=8,
        l2_leaf_reg=3.0, random_seed=400 + fold_seed,
        verbose=0, eval_metric='AUC', allow_writing_files=False,
        early_stopping_rounds=150, task_type='GPU', devices='0',
    )))

    return models

model_names = [name for name, _ in get_models(0)]
N_MODELS = len(model_names)
print(f"Models ({N_MODELS}): {model_names}")
```

```python
# 5. TRAINING LOOP -> 5-Fold Stratified CV
skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=SEED)

oof_preds = np.zeros((len(X), N_MODELS), dtype=np.float32)
test_preds = np.zeros((len(X_test), N_MODELS), dtype=np.float32)

for fold, (tr_idx, va_idx) in enumerate(skf.split(X, y)):
    X_tr, X_va = X[tr_idx], X[va_idx]
    y_tr, y_va = y[tr_idx], y[va_idx]
    print(f"\n{'='*60}")
    print(f"FOLD {fold+1}/{N_FOLDS}: train={len(X_tr)} (pos={y_tr.sum()}), val={len(X_va)} (pos={y_va.sum()})")
    print(f"{'='*60}")

    fold_models = get_models(fold)

    for i, (name, model) in enumerate(fold_models):
        print(f"  Training {name}...", end=" ", flush=True)

        if name.startswith('XGB'):
            model.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False)
            print(f"best_iter={model.best_iteration}", end=" ")
        elif name.startswith('LGB'):
            if name == 'LGB_goss':
                model.fit(X_tr, y_tr)
            else:
                model.fit(X_tr, y_tr, eval_set=[(X_va, y_va)],
                          callbacks=[
                              __import__('lightgbm').early_stopping(150, verbose=False),
                              __import__('lightgbm').log_evaluation(0),
                          ])
                print(f"best_iter={model.best_iteration_}", end=" ")
        elif name == 'CatBoost':
            model.fit(X_tr, y_tr, eval_set=(X_va, y_va))
            print(f"best_iter={model.best_iteration_}", end=" ")

        proba = model.predict_proba(X_va)[:, 1]
        oof_preds[va_idx, i] = proba
        test_preds[:, i] += model.predict_proba(X_test)[:, 1] / N_FOLDS

        # Quick F1 check
        best_f1 = max(f1_score(y_va, (proba > t).astype(int), zero_division=0)
                      for t in np.arange(0.1, 0.9, 0.02))
        print(f"val_F1={best_f1:.4f}")

        del model; gc.collect()
```

```python
# 6. INDIVIDUAL MODEL PERFORMANCE
for i, name in enumerate(model_names):
    best_f1, best_t = 0, 0.5
    for t in np.arange(0.05, 0.95, 0.005):
        f1 = f1_score(y, (oof_preds[:, i] > t).astype(int), zero_division=0)
        if f1 > best_f1:
            best_f1 = f1; best_t = t
    print(f"  {name:12s}: OOF F1={best_f1:.4f} @ threshold={best_t:.3f}")
```

```python
# 7. STACKING META-LEARNER
# Meta-learner 1: Logistic Regression
meta_lr = LogisticRegression(C=1.0, class_weight='balanced', max_iter=1000, random_state=SEED)
meta_lr.fit(oof_preds, y)
meta_oof_lr = meta_lr.predict_proba(oof_preds)[:, 1]
meta_test_lr = meta_lr.predict_proba(test_preds)[:, 1]
print(f"  LR coefficients: {meta_lr.coef_[0]}")

if FAST_MODE:
    meta_oof = meta_oof_lr
    meta_test = meta_test_lr
else:
    meta_xgb = XGBClassifier(
        n_estimators=200, max_depth=3, learning_rate=0.05,
        scale_pos_weight=scale_w * 0.5, random_state=SEED,
        tree_method='hist', device='cuda', max_bin=256,
        eval_metric='aucpr',
    )
    meta_xgb.fit(oof_preds, y)
    meta_oof_xgb = meta_xgb.predict_proba(oof_preds)[:, 1]
    meta_test_xgb = meta_xgb.predict_proba(test_preds)[:, 1]
    meta_oof = 0.5 * meta_oof_lr + 0.5 * meta_oof_xgb
    meta_test = 0.5 * meta_test_lr + 0.5 * meta_test_xgb

print("\n  Searching simple equal-weight blend baseline...")
equal_blend_oof = oof_preds.mean(axis=1)
equal_blend_test = test_preds.mean(axis=1)
best_blend_f1, best_blend_t = 0, 0.5
for t in np.arange(0.05, 0.95, 0.005):
    f1 = f1_score(y, (equal_blend_oof > t).astype(int), zero_division=0)
    if f1 > best_blend_f1:
        best_blend_f1 = f1
        best_blend_t = t
print(f"  Equal blend: F1={best_blend_f1:.4f} @ threshold={best_blend_t:.3f}")

# Find best stacked threshold
best_stack_f1, best_stack_t = 0, 0.5
for t in np.arange(0.05, 0.95, 0.005):
    f1 = f1_score(y, (meta_oof > t).astype(int), zero_division=0)
    if f1 > best_stack_f1:
        best_stack_f1 = f1; best_stack_t = t
print(f"  Stacked:    F1={best_stack_f1:.4f} @ threshold={best_stack_t:.3f}")

# Choose best approach
if best_stack_f1 >= best_blend_f1:
    print(f"\n  Using STACKED (F1={best_stack_f1:.4f})")
    final_oof = meta_oof
    final_test = meta_test
    final_threshold = best_stack_t
    final_f1 = best_stack_f1
else:
    print(f"\n  Using BLENDED (F1={best_blend_f1:.4f})")
    final_oof = equal_blend_oof
    final_test = equal_blend_test
    final_threshold = best_blend_t
    final_f1 = best_blend_f1
```

```python
# 8. POST-PROCESSING (Rule-based corrections)
oof_binary = (final_oof > final_threshold).astype(int)
test_binary = (final_test > final_threshold).astype(int)

# Rule 1: Force certain anomaly zones to positive
# (X3_ord >= 60 OR X4_ord >= 64 OR (X3_ord >= 50 AND X4_ord >= 30))
# These have 90-100% anomaly rate in training data
oof_forced_pos = train_zone_certain.sum() - oof_binary[train_zone_certain == 1].sum()
test_forced_pos = test_zone_certain.sum() - test_binary[test_zone_certain == 1].sum()
oof_binary[train_zone_certain == 1] = 1
test_binary[test_zone_certain == 1] = 1
print(f"  Rule 1 (force positive): flipped {oof_forced_pos} train, {test_forced_pos} test")

# Rule 2: Force safe zones to negative (CONSERVATIVE -> only apply if model is uncertain)
# (X3_ord == 0 AND X4_ord == 0 AND X5_ord <= 1 AND model probability < 0.3)
safe_mask_oof = (train_zone_safe == 1) & (final_oof < 0.3)
safe_mask_test = (test_zone_safe == 1) & (final_test < 0.3)
oof_forced_neg = oof_binary[safe_mask_oof].sum()
test_forced_neg = test_binary[safe_mask_test].sum()
oof_binary[safe_mask_oof] = 0
test_binary[safe_mask_test] = 0
print(f"  Rule 2 (force negative): flipped {oof_forced_neg} train, {test_forced_neg} test")

# Check post-processed OOF F1
pp_f1 = f1_score(y, oof_binary)
print(f"\n  Post-processed OOF F1: {pp_f1:.4f} (was {final_f1:.4f})")
```

```python
# 9. FINAL RESULTS
print(classification_report(y, oof_binary))
print(f"Test: 0={int((test_binary==0).sum())}, 1={int((test_binary==1).sum())}, rate={test_binary.mean():.4f}")
```
